## Label - Price (2019 Dollars)

```python
print(data.price.describe())

data.price.hist(bins=500)
plt.title("Price ($) Histogram", weight='bold', fontsize=16 )
plt.ylabel('Number of Listings', fontsize=14)
plt.xlabel('Price ($)', fontsize=14)
plt.axvline(x=data.price.median(),
            color='orange',
            linestyle='--',
            label='median price')
plt.axvline(x=data.price.quantile(0.95),
            color='red', 
            linestyle='--', 
            label='95% percentile')
plt.legend(loc='upper right')
plt.xlim(0, 1000)
plt.show()
```

### Price Histogram Notes:
1. Average price is 152 dollars 
2. Cheapest listing - 0 dollars (why is this??) and most expensive - 10000 dollars - ((is this a long stay??)
3. Most values fall in < 200 dollars range with a long tail (should we clip this tail?)
4. Strong left-side skew, we need to normalize the price to be gaussian (normal) dist 
5. Do we also want the price to be scaled to 0-1 also?? 

```python
## have to remove 0 value or we create an infinite series log(0) == -inf 
## or set it to zero
data['price'] = np.where(data.price==0, 1, data.price)
data['price_norm'] = np.log(data.price)
sns.distplot(data.price_norm, bins=25)
plt.xlabel("Price (post-normalization)", fontsize=14)
plt.ylabel("Share of Viewings", fontsize=14)
plt.title("Normalized Price Histogram", fontsize=16, weight='bold')
plt.show()
```

### Autocorrelation between features

```python
data.corr()
```

```python
## Focusing on high (abs) corr values
correlation_df = data.corr().unstack().reset_index()
correlation_df.columns = ['feature_1', 'feature_2', 'corr_coeff']
correlation_df = correlation_df[correlation_df !=1] # all features are perfect corr to themselves

CORRELATION_THRESHOLD = 0.20 # absolute value so captures +- corr
correlation_df[correlation_df.corr_coeff.abs() > CORRELATION_THRESHOLD]
```

```python
# Host-id and Id are heavily correlated (makes sense) and so is reviews per month and number reviews. 
corr = data.corr()
sns.heatmap(corr)
```

### Correlation with label 

```python
correlation_df = data.corr().unstack().reset_index()
correlation_df.columns = ['feature_1', 'feature_2', 'corr_coeff']

price_corr = correlation_df[(correlation_df.feature_1=="price") & 
                            ~(correlation_df.feature_2.str.contains('price'))].copy()
price_corr['corr_coeff_abs'] = price_corr.corr_coeff.abs()
price_corr.sort_values('corr_coeff_abs', ascending=False)
```
